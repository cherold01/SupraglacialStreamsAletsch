```julia
p = Plots.plot() #title = "Normalised values of different Stream properties"
ones = [1 for _ in 1:15]
twos = [2 for _ in 1:15]
threes = [3 for _ in 1:15]
fours = [4 for _ in 1:15]
fives = [5 for _ in 1:15]
sixes = [6 for _ in 1:15]
categories = ["Q_max", "Width", "Depth", "f", "v", "Gradient"]
Plots.scatter!(ones, mean_object(polyg.Q_max_meas)/maximum(mean_object(polyg.Q_max_meas)), c = colors[polyg.Stream], label = false)
Plots.scatter!(twos, mean_object(polyg.W_avg_measured)/maximum(mean_object(polyg.W_avg_measured)), c = colors[polyg.Stream], label = false)
Plots.scatter!(threes, mean_object(polyg.Depth_avg)/maximum(mean_object(polyg.Depth_avg)),c = colors[polyg.Stream], label = false)
Plots.scatter!(fours, mean_object(polyg.f_Qmax)/maximum(mean_object(polyg.f_Qmax)),c = colors[polyg.Stream], label = false)
Plots.scatter!(fives, mean_object(polyg.v_Qmax)/maximum(mean_object(polyg.v_Qmax)),c = colors[polyg.Stream], label = false)
Plots.scatter!(sixes, mean_object(polyg.Grad_straight)/maximum(mean_object(polyg.Grad_straight)),c = colors[polyg.Stream], label = false)
Plots.xlims!(0, length(categories)+1)
Plots.ylims!(0,1.05)
Plots.xticks!(1:length(categories), categories)

Plots.plot!(1,1, label="Stream 1", marker=:circle, color=colors[1], linealpha = 0)
Plots.plot!(1,1, label="Stream 2", marker=:circle, color=colors[2], linealpha = 0)
Plots.plot!(1,1, label="Stream 3", marker=:circle, color=colors[3], linealpha = 0)
Plots.plot!(1,1, label="Stream 4", marker=:circle, color=colors[4], linealpha = 0)
Plots.plot!(1,1, label="Stream 5", marker=:circle, color=colors[5], linealpha = 0)
display(p)
#savefig("../data/products/plots/final/Normalised-values_final.png")
```
