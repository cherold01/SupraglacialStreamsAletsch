```julia
p = Plots.plot()
ones = [1 for _ in 1:15]
twos = [2 for _ in 1:15]
threes = [3 for _ in 1:15]
categories = ["Q_max", "Width", "Depth"]

mean_Q_max = mean_object(polyg.Q_max_meas)
mean_Width = mean_object(polyg.W_avg_measured)
mean_Depth = mean_object(polyg.Depth_avg)

# Create subplots
p1 = Plots.scatter(ones, mean_Q_max, c=colors[polyg.Stream], label=false, xlabel="Categories", ylabel="Q_max")
p2 = Plots.scatter(twos, mean_Width, c=colors[polyg.Stream], label=false, xlabel="Categories", ylabel="Width")
p3 = Plots.scatter(threes, mean_Depth, c=colors[polyg.Stream], label=false, xlabel="Categories", ylabel="Depth")

# Combine subplots
p = Plots.plot(p1, p2, p3, layout=(3, 1), link=:x, xticks=(1:3, categories))

#Plots.xlims!(0, 4)

Plots.plot!(1,1, label="Stream 1", marker=:circle, color=:red, linealpha = 0, markerstrokecolor=:red)
Plots.plot!(1,1, label="Stream 2", marker=:circle, color=:blue, linealpha = 0, markerstrokecolor=:blue)
Plots.plot!(1,1, label="Stream 3", marker=:circle, color=:green, linealpha = 0, markerstrokecolor=:green)
Plots.plot!(1,1, label="Stream 4", marker=:circle, color=:darkorange, linealpha = 0, markerstrokecolor=:darkorange)
Plots.plot!(1,1, label="Stream 5", marker=:circle, color=:pink, linealpha = 0, markerstrokecolor=:pink)
display(p)
```

```julia
# Sample data for 5 categories with 3 data points each
categories = ["Category 1", "Category 2", "Category 3", "Category 4", "Category 5"]
data = [
    [10, 12, 14],
    [22, 23, 21],
    [30, 28, 32],
    [40, 42, 39],
    [50, 48, 52]
]

# Create the plot
f = Plots.plot()
for (i, category) in enumerate(categories)
    scatter!(fill(category, length(data[i])), data[i], label=category)
end

# Set labels and title
xlabel!("Categories")
ylabel!("Values")
title!("Differences between Categories")

# Display the plot
display(f)
```
